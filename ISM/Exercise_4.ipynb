```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
k = 1.38e-23  # Boltzmann constant (J/K)
G = 6.67e-11  # Gravitational constant (N m^2/kg^2)
m_H = 1.67e-27  # Mass of a hydrogen atom (kg)

# density (particles/m^3)
n_SI = 1e-11     

# Calculate mass density (kg/m^3)
rho = n_SI * m_H

# Range of temperatures (K)
T = np.logspace(0, 3, 200)  # 10 to 1000 K

M_J=4*(k*T/G/m_H)**1.5*rho**(-0.5)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(T, M_J)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Temperatur (K)', fontsize=14)
plt.ylabel('Jeans mass (kg)', fontsize=14)
plt.title('Jeans mass vs. Temperature (n = 1e-11/m$^3$)', fontsize=16)
# plt.grid(True)
plt.show()
```
